```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split
from keras.layers import Input
from keras.layers import Dense
from keras.optimizers import Adam
from keras.models import Sequential
import numpy as np
import keras
from collections import Counter
```

```python
def wrangle_data(data, bool_X_test = 0):

    if bool_X_test == 0:
        
        features = data.drop(columns=['loan_status', 'id'])
        labels = data.loan_status
        
        x_train, x_test, y_train, y_test = train_test_split(features, labels, test_size=0.3)
        # Select numeric column names
        numeric_cols = x_train.select_dtypes(include=['float64', 'int64']).columns
        # Create new standard scaler object
        scaler = StandardScaler()
        x_train[numeric_cols] = scaler.fit_transform(x_train[numeric_cols])
        x_test[numeric_cols] = scaler.transform(x_test[numeric_cols])
        
        x_train = pd.get_dummies(x_train, drop_first=True, dtype='int')
        x_test = pd.get_dummies(x_test, drop_first=True, dtype='int')
        
        return(x_train, x_test, y_train, y_test)
        
    else:
        y_labels = data.id
        X_test = data.drop(columns=['id'])
        
        # Select numeric column names
        numeric_cols = X_test.select_dtypes(include=['float64', 'int64']).columns
        # Create new standard scaler object
        scaler = StandardScaler()
        X_test[numeric_cols] = scaler.fit_transform(X_test[numeric_cols])
        X_test = pd.get_dummies(X_test, drop_first=True, dtype='int')

        return X_test, y_labels
```

```python
data = pd.read_csv('train.csv')
x_train, x_test, y_train, y_test = wrangle_data(data, 0)
x_train.head()
```

```python
model = Sequential()

model.add(Input(shape=(x_train.shape[1], )))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
opt = Adam(learning_rate = 0.001)
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=10, verbose=1, batch_size=1)
```

```python
data = pd.read_csv('test.csv')
X_test, y_labels = wrangle_data(data, 1)
X_test.head()
```

```python
predictions = []
y_estimate = model.predict(X_test)
for y in y_estimate:
    if y >= 0.5:
        predictions.append(1)
    else:
        predictions.append(0)

output_df = pd.DataFrame({'id': y_labels, 'loan_status': predictions})
output_df.head()
```

```python
output_df.to_csv('submission.csv')
```
